````python
from numpy.core.memmap import dtype

import copy
import random
import time
import numpy as np
class OperacionesMatriciales:
    def __init__(self, M):
        self.M = M
    def nueva_matriz(self,N,i):
        arr = copy.deepcopy(N)
        if (len(arr) == 2):
            return arr
        else:
            #np.delete(arr,0,axis = 0)
            arr.pop(0)
            for j in arr:
                j.pop(i)
                #np.delete(arr,j,axis=1)
            return arr
        
    def determinat(self,N):
        if (len(N) == 1):
            return N[0]
        elif len(N) == 2:
            return N[0][0]*N[1][1] - N[0][1]*N[1][0]
        
        else:
            pro = 0
            for i in range(len(N[0])):
                pro += ((-1)**i*N[0][i]*self.determinat(self.nueva_matriz(N,i)))
            return pro
            
    def trazas(self):
        return [self.M[i][i] for i in range(len(self.M))], sum([self.M[i][i] for i in range(len(self.M))])

    def transpuestaM(self,M):
        N = np.zeros_like(M)
        #print(N)
        for i in range(len(M)):
            for j in range(len(M[0])):
                N[i][j] = M[j][i]
        return N
                

    def propiedadCiclicaTraza(self):
        t = 0
        for i in range(len(self.M[0])):
            for j in range(len(self.M[1])):
                t += self.M[i][j]**2
        return t

    def sumaMatrices(self,A):
        m = int(len(self.M[1]))
        n = int(len(A))
        if (m == n):
            suma = np.zeros((m,n))
            #print(suma)
            for i in range(n):
                for j in range(m):
                    if (j < len(A[0])):
                        suma[i][j] = self.M[i][j] + A[i][j]
                    else:
                        suma[i][j] = self.M[i][j]
            return suma
        else:
            print('no se pueden sumar')

    def restaMatrices(self,A):
        m = int(len(self.M[1]))
        n = int(len(A))
        if (m == n):
            resta = np.zeros((m,n))
            #print(suma)
            for i in range(n):
                for j in range(m):
                    if (j < len(A[0])):
                        resta[i][j] = self.M[i][j] - A[i][j]
                    else:
                        resta[i][j] = -self.M[i][j]
            return resta
        else:
            print('no se pueden restar')
    
    def multiesc(self,num):
        mult = np.zeros_like(self.M)
        for i in range(len(self.M)):
            for j in range(len(self.M[1])):
                mult[i][j] = num*self.M[i][j]
        return mult

    def multimatrices(self,B):
        m = int(len(self.M[1]))
        n = int(len(B))
        if (m == n):
            multiplicacion = np.zeros((len(self.M),len(B[1])))
            for i in range(len(self.M)):
                for j in range(len(B[1])):
                    val = 0
                    for k in range(len(self.M[0])):
                        val += self.M[i][k]*B[k][j] 
                    multiplicacion[i][j] = val
            return multiplicacion

        else:
            print('no se puede multiplicar')


    def bloqueM2x2(self):
        if len(self.M[0]) == 4 and len(self.M[1]) == 4:
            M = np.array(self.M)
            a = np.split(np.hsplit(M,2)[0],2)[0]
            b = np.split(np.hsplit(M,2)[1],2)[0]
            c = np.split(np.hsplit(M,2)[0],2)[1]
            d = np.split(np.hsplit(M,2)[1],2)[1]
            #a = np.split(np.hsplit(M,2)[0],2), np.split(np.hsplit(M,2)[1],2)
            return a,b,c,d
        else:
            return None
    
    def vectorMatrixF(self,M):
        n = len(M)
        vectM = np.zeros((n*n))
        for i in range(n):
            for j in range(n):
                vectM[i*n+j] = M[i][j]     
        return vectM,n

    def vectorMatrixC(self,M):
        n = len(M[0])
        vectM = np.zeros((n*n))
        for i in range(n):
            for j in range(n):
                vectM[j*n+i] = M[i][j]     
        return vectM,n

    def sumavecmat(self,B,M):
        B,n = self.vectorMatrixC(B)
        M,m = self.vectorMatrixC(M)
        c = np.zeros_like(M)
        for i in range(len(c)):
            c[i] = M[i] + B[i]
        return c
        
    def multvecmat(self,A,B):
        A,m = self.vectorMatrixF(A)
        print(A)
        B,n = self.vectorMatrixC(B)
        print(B)
        ma = np.zeros((m*n))
        #i = 0
        if n == m:
            #while (i<=m):
            
            for i  in range(n):
                for k in range(n):
                    for j in range(m):
                        ma[i*n+k] += A[n*i+j]*B[k*n+j] 
                    #print(c)
                    
            return ma
        else:
            print('no se puede')
            return None


    def dispM(self,M):
        dispersa = []
        for i in range(len(M)):
            for j in range(len(M[0])):
                if M[i][j] != 0:
                    dispersa.append([i,j,M[i][j]])
        return dispersa

    def transM(self,N):
        for index,value in enumerate(N):
            '''print(index) indexacion
                print(value[0]) fila i
                print(value[1]) columna j
                print(value[2]) valor w
                print(j)'''
            tmp = value[0]
            value[0] = value[1]
            value[1] = tmp
            
        return N
    def trazaMdisp(self,N):
        traza = 0
        for index,value in enumerate(N):
            #for j in range(1):
            if value[0] == value[1]:
                #if index == value[j] and index == value[j]:
                #print(index, value[0],index, value[1])
                traza += value[2]
        return traza

    def norm(self,V):
        normv = 0
       
        value, n = V
        #print(n)
        #print(value)
        #print(n[0]**2+n[1]**2)
        for i in range(len(value)):
            normv += value[i]**2
        return np.sqrt(normv)

    def normP(self,V,p):
        normv = 0
       
        value, n = V
        #print(n)
        #print(value)
        #print(n[0]**2+n[1]**2)
        for i in range(len(value)):
            normv += value[i]**p
        return (normv)**(1/p)

    def norml1(self,M):
        fila = np.zeros((len(M)))
        for i,j in enumerate(M):
            fila[i] = sum(j)
        return max(fila)

    def normlinf(self,M):
        n = len(M[0])
        col = np.zeros(n)
        M = self.transpuestaM(M)
        for i,j in enumerate(M):
            col[i] = sum(j)
            print(i)
            #print(j*n)
            print(col)
        return max(col)

def main():
    #matriz = [[1,2,4],[3,4,7],[5,6,7]]
    matriz = np.random.randint(-5,20,(4,4))
    M = matriz.tolist()
    print(matriz)
    objOpM = OperacionesMatriciales(M)
    '''print(objOpM.determinat(M))
    #print(round(np.linalg.det(matriz)))
    print(objOpM.trazas())
    print(objOpM.propiedadCiclicaTraza())
    matriz2 = np.random.randint(-10,10,(8,5))
    print(objOpM.sumaMatrices(matriz2))
    print(objOpM.restaMatrices(matriz2))
    print(objOpM.multiesc(2))
    print(objOpM.multimatrices(matriz2))'''
    #A,B,C,D = objOpM.bloqueM2x2()
    #print(objOpM.vectorMatrixF())
    #matriz2 = np.random.randint(-10,10,(4,4))
    matriz_transpuesta = objOpM.transpuestaM(matriz)
    '''print(matriz)
    print(objOpM.vectorMatrixF(matriz))
    #print(objOpM.vectorMatrixC(matriz2))
    #print(objOpM.sumavecmat(matriz,matriz2))
    #print('multiplicacion normal\n',objOpM.multimatrices(matriz2))
    #print('multiplicacion vectorial:',objOpM.multvecmat(matriz,matriz2))
    #matrices_dispersa =  objOpM.dispM()
    #print('matriz dispersa', matrices_dispersa)
    #print(matrices_dispersa[(1,2)])'''
    #print('matriz dispersa transpuesta', objOpM.tranDispM(matrices_dispersa))
    m_d1 =  objOpM.dispM(matriz)
    #print(matrices_dispersa)
    #print(objOpM.transM(m_d1))
    #print(objOpM.trazaMdisp(m_d1))
    #vector = objOpM.vectorMatrixF(matriz) #return el array y el tamaño de la fila
    #print(objOpM.norm(vector)) 
    #print(objOpM.normP(vector,3))
    #objOpM = OperacionesMatriciales(m_d1)
    print(objOpM.norml1(m_d1))
    print(objOpM.norml1(matriz_transpuesta))

if __name__ == '__main__':
    main()            
````
